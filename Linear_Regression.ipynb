```python
with open('Data/gases.csv', newline='') as csvfile:
    gases = csv.DictReader(csvfile, delimiter=',', quotechar='\'', quoting=csv.QUOTE_NONNUMERIC)
#delimiter specifies the character used to separate each field. The default is the comma (',').
#quotechar specifies the character used to surround fields that contain the delimiter character. The default is a double quote (' " ').
    NO2 = []
    NO2T = []
    NO2TH = []
    NO2_we = []
    NO2_ae = []
    NO2_mo = []

    SO2 = []
    SO2T = []
    SO2TH = []
    SO2_we = []
    SO2_ae = []
    SO2_mo = []

    CO = []
    COT = []
    COTH = []
    CO_we = []
    CO_ae = []
    CO_mo = []

    O3 = []
    O3T = []
    O3TH = []
    O3_we = []
    O3_ae = []
    O3_mo = []

    O3NO2 = []

    G = []

    T = []
    H = []
    for row in gases:

        NO2 = np.append(NO2, [row['NO2_we'], row['NO2_ae']])
        NO2T = np.append(NO2T, [row['NO2_we'], row['NO2_ae'], row['Temperatura']])
        NO2TH = np.append(NO2TH, [row['NO2_we'], row['NO2_ae'], row['Temperatura'], row['Humedad']])
        NO2_mo = np.append(NO2_mo, row['NO2_Mompean'])
        NO2_we.append(row['NO2_we'])
        NO2_ae.append(row['NO2_ae'])

        SO2 = np.append(SO2, [row['SO2_WE'], row['SO2_AE']])
        SO2T = np.append(SO2T, [row['SO2_WE'], row['SO2_AE'], row['Temperatura']])
        SO2TH = np.append(SO2TH, [row['SO2_WE'], row['SO2_AE'], row['Temperatura'], row['Humedad']])
        SO2_mo = np.append(SO2_mo, row['SO2_Mompean'])
        SO2_we.append(row['SO2_WE'])
        SO2_ae.append(row['SO2_AE'])

        CO = np.append(CO, [row['CO_we'], row['CO_ae']])
        COT = np.append(COT, [row['CO_we'], row['CO_ae'], row['Temperatura']])
        COTH = np.append(COTH, [row['CO_we'], row['CO_ae'], row['Temperatura'], row['Humedad']])
        CO_mo = np.append(CO_mo, row['CO_Mompean'])
        CO_we.append(row['CO_we'])
        CO_ae.append(row['CO_ae'])

        O3 = np.append(O3, [row['O3_we'], row['O3_ae']])
        O3T = np.append(O3T, [row['O3_we'], row['O3_ae'], row['Temperatura']])
        O3TH = np.append(O3TH, [row['O3_we'], row['O3_ae'], row['Temperatura'], row['Humedad']])
        O3_mo = np.append(O3_mo, row['O3_Mompean'])
        O3_we.append(row['O3_we'])
        O3_ae.append(row['O3_ae'])

        O3NO2 = np.append(O3NO2, [row['O3_we'], row['O3_ae'], row['NO2_we'], row['NO2_ae'], row['Temperatura'], row['Humedad']])

        G = np.append(G, [row['O3_we'], row['O3_ae'], row['NO2_we'], row['NO2_ae'], row['SO2_WE'], row['SO2_AE'], 
                              row['CO_we'], row['CO_ae'], row['Temperatura'], row['Humedad']])

        T.append(row['Temperatura'])
        H.append(row['Humedad'])
    
    NO2 = np.reshape(NO2, (-1, 2))
    NO2T = np.reshape(NO2T, (-1, 3))
    NO2TH = np.reshape(NO2TH, (-1, 4))

    SO2 = np.reshape(SO2, (-1, 2))
    SO2T = np.reshape(SO2T, (-1, 3))
    SO2TH = np.reshape(SO2TH, (-1, 4))

    CO = np.reshape(CO, (-1, 2))
    COT = np.reshape(COT, (-1, 3))
    COTH = np.reshape(COTH, (-1, 4))

    O3 = np.reshape(O3, (-1, 2))
    O3T = np.reshape(O3T, (-1, 3))
    O3TH = np.reshape(O3TH, (-1, 4))

    O3NO2 = np.reshape(O3NO2, (-1, 6))

    G = np.reshape(G, (-1, 10))

    NO2_prueba = []
    O3NO2_prueba = []
    SO2_prueba = []
    CO_prueba = []
    G_prueba = []
```
